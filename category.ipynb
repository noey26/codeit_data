```python
import pandas as pd
```

```python
df = pd.read_csv('codeit/laptops.csv')
df.head()
```

```python
brand_nation = {
    'Dell': 'U.S.',
    'Apple': 'U.S.',
    'Acer': 'Taiwan',
    'HP': 'U.S.',
    'Lenovo': 'China',
    'Alienware': 'U.S.',
    'Microsoft': 'U.S.',
    'Asus': 'Taiwan'
}
```

```python
df['brand'].map(brand_nation)
```

```python
df['brand_nation'] = df['brand'].map(brand_nation)
```

```python
df
```

```python
nation_groups = df.groupby('brand_nation')
type(nation_groups)
```
